```python
""" 01. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list"""

def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

numbers = [2,44,56,3,8,9,1,0,10,12,1,4,5,3,5]
print("Given list are:" , numbers)
print("your all even numbers from list: ",[num for num in numbers if num % 2 == 0])
print("Sum of all even numbers is: ", sum_even_numbers(numbers))
```

```python
# 04. Write a Python function that checks if a given number is prime or not from 1 to 200

n = int(input("Enter any number between 1 and 200: "))

def is_prime(n):
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

if is_prime(n):
    print(f"{is_prime(n)} : {n} is a prime number")
else:
      print(f"{is_prime(n)} : {n} is NOT a prime number")
```

```python
""" 05. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms"""

def fib_seq(n):
    a, b = 0, 1
    for i in range(n):
        print(a)
        a, b = b, a + b
print(fib_seq(10))
```

```python
# 06. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

print(list(powers_of_two(4)))

```

```python
# 07. Implement a generator function that reads a file line by line and yields each line as a string

def read_file(file_path):
    with open(file_path) as f:
        for line in f:
            yield line.strip()

# Example usage
for line in read_file('example.txt'):
    print(line)

# Output:
# Hello
# World
# Python
```

```python
# 08. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

def sort_by_second(tuples):
    return sorted(tuples, key=lambda x: x[1])

tuples = [(5, 3), (2, 1), (4, 2),(8,9)]
print(f"Given list are: {tuples}")
print(f"sorted list based on the second element : {sort_by_second(tuples)}")
```

```python
# 09. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

def cel_to_fah(celsius_list):
    return list(map(lambda c: (c * 9/5) + 32, celsius_list))

celsius_list = [0, 20, 37, 100]
print(f"Given list of temperatures in Celsius: {celsius_list} ")
print(f"after converting temperatures into Fahrenheit: {cel_to_fah(celsius_list)}")
```

```python
# 10. Create a Python program that uses filter() to remove all the vowels from a given string

def remove_vowels(string1):
    return ''.join(filter(lambda ch: ch.lower() not in 'aeiou', string1))

string1 = "Hello World"
print("Given string is : ", string1)
print("After remove Vowels from string: ",remove_vowels(string1))
```
